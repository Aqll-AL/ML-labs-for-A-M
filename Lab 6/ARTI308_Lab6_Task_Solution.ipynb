# Lab 6 Task — Linear Regression on Ecommerce Customers

Applying what was learned in Lab 6 to a new dataset: **Ecommerce Customers**.

The goal is to build a Linear Regression model that predicts **Yearly Amount Spent** by a customer based on their usage behaviour.

## Step 1: Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

## Step 2: Load the Dataset

```python
ecom = pd.read_csv('Ecommerce_Customers')
```

## Step 3: Explore the Data

```python
ecom.head()
```

```python
ecom.info()
```

```python
ecom.describe()
```

## Step 4: Basic Data Cleaning — Check for Missing Values

```python
print('Missing values per column:')
print(ecom.isnull().sum())
```

## Step 5: EDA

### Pairplot of Numeric Features

```python
sns.pairplot(ecom, vars=['Avg. Session Length', 'Time on App',
                         'Time on Website', 'Length of Membership',
                         'Yearly Amount Spent'])
plt.show()
```

### Correlation Heatmap

```python
sns.heatmap(ecom.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

## Step 6: Prepare Data for Modeling

Drop non-numeric columns (Email, Address, Avatar). Predict `Yearly Amount Spent`.

```python
X = ecom[['Avg. Session Length', 'Time on App',
          'Time on Website', 'Length of Membership']]
y = ecom['Yearly Amount Spent']

print('Features shape:', X.shape)
print('Target shape:  ', y.shape)
```

## Step 7: Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=101
)
print('Train size:', X_train.shape)
print('Test size: ', X_test.shape)
```

## Step 8: Train the Linear Regression Model

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

### Model Coefficients

```python
print('Intercept:', lm.intercept_)
coeff_df = pd.DataFrame(lm.coef_, X.columns, columns=['Coefficient'])
coeff_df
```

**Interpreting the coefficients:**

- Holding all other features fixed, a 1-minute increase in **Avg. Session Length** is associated with an increase of ~\$25–26 in yearly spending.
- Holding all other features fixed, a 1-minute increase in **Time on App** is associated with an increase of ~\$38–39 — the strongest per-minute driver.
- Holding all other features fixed, a 1-minute increase in **Time on Website** has a very small effect, suggesting the website is less influential than the app.
- Holding all other features fixed, each additional year of **Length of Membership** is associated with an increase of ~\$61–62 — the largest overall driver.

## Step 9: Predictions

```python
predictions = lm.predict(X_test)

plt.scatter(y_test, predictions, alpha=0.5)
plt.xlabel('Actual Yearly Amount Spent')
plt.ylabel('Predicted Yearly Amount Spent')
plt.title('Actual vs. Predicted')
plt.show()
```

### Residual Distribution

```python
sns.histplot((y_test - predictions), bins=40, kde=True)
plt.title('Residuals Distribution')
plt.xlabel('Residual')
plt.show()
```

## Step 10: Evaluate Model Performance

```python
mae  = metrics.mean_absolute_error(y_test, predictions)
mse  = metrics.mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)

print(f'MAE:  {mae:.2f}')
print(f'MSE:  {mse:.2f}')
print(f'RMSE: {rmse:.2f}')
```

## Summary

The Linear Regression model performs very well on the Ecommerce Customers dataset:

- The residuals are approximately normally distributed and centered near zero, indicating a good fit.
- The low RMSE (relative to the scale of `Yearly Amount Spent`) confirms the model's predictive accuracy.
- **Length of Membership** and **Time on App** are the most impactful features for predicting yearly spending.
