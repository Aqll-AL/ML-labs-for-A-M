```python
plt.figure(figsize=(10,10))
sns.countplot(df['diagnosis'])
plt.title('Diagnosis Distribution')
plt.show()
```

M (Malignant): Indicates a cancerous tumor.
B (Benign): Indicates a non-cancerous tumor.


*   Shows that non-cancerous tumor is more that cancerous



```python
plt.figure(figsize=(8,5))
sns.histplot(df['radius_mean'])
plt.title('radius_mean')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['radius_worst'])
plt.title('Distribution of radius_worst')
plt.xlabel('radius_worst')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df,x='radius_mean',y='radius_worst')
plt.title('Scatterplot of radius mean vs radius worst')
plt.show()
```

* Strong positive relationship

```python
malignant_df = df[df['diagnosis'] == 'M']

plt.figure(figsize=(8, 6))
sns.histplot(malignant_df['radius_mean'])
plt.title('Distribution of Radius Mean for Malignant Tumors')
plt.xlabel('Radius Mean')
plt.ylabel('Frequency')
plt.show()
```

* Malignant tumors tend to have larger radius mean

```python
Benign_df = df[df['diagnosis'] == 'B']

plt.figure(figsize=(8, 6))
sns.histplot(Benign_df['radius_mean'])
plt.title('Distribution of Radius Mean for benign Tumors')
plt.xlabel('Radius Mean')
plt.ylabel('Frequency')
plt.show()
```

* Benign tumors cluster at smaller values

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['radius_worst','radius_mean']].corr(),annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

* High correlation between radius mean & radius wors

```python
sns.boxplot(x='diagnosis', y='radius_mean', data=df)
plt.title("Radius Mean by Diagnosis")
```

* Malignant tumors usually have higher median values
* Shows outliers clearly

```python
sns.boxplot(x='diagnosis', y='texture_mean', data=df)
plt.title('Texture Mean by Diagnosis')
```

```python
sns.pairplot(df[['diagnosis','radius_mean','texture_mean','perimeter_mean','area_mean']], hue='diagnosis')
```

* Shows separation between benign & malignant
